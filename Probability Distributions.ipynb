# Continuous Uniform Distribution

$\begin{align}
\textrm{PDF}(X) = \frac{1}{b-a}\quad \text{for } a \leq x \leq b, 0 \text{ otherwise}
\end{align}$

$\begin{align}
\textrm{CDF}(X) = \frac{x-a}{b-a}\quad \text{for } a \leq x \leq b, 0 \text{ otherwise}
\end{align}$

$\begin{align}
\textrm{CDF}^{-1}(X) = x(b - a) + a\quad \text{for } 0 \leq x \leq 1
\end{align}$

# Exponential Distribution

$\begin{align}
\textrm{PDF}(X) = \lambda\ e^{-\lambda x}\quad \text{for } x \geq 0
\end{align}$

$\begin{align}
\textrm{CDF}(X) = -e^{-\lambda x} + 1\quad \text{for } x \geq 0
\end{align}$

$\begin{align}
\textrm{CDF}^{-1}(X) = -\frac{\textrm{ln}(1-x)}{\lambda}\quad \text{for } 0 \leq x \leq 1
\end{align}$
